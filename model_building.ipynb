```python
import pandas as pd
```

```python
from sqlalchemy import create_engine
engine=create_engine('postgresql://postgres:postgres@127.0.0.1:5432/used_cars')
```

```python
df=pd.read_sql('select * from used_cars', engine)
df.head()
```

```python
y=df[['price_eur']]
X_cat=df[['maker']]
X_numerical=df[['mileage']]
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
ohe.fit(X_cat[['maker']])
X_cat_transformed=pd.DataFrame(ohe.transform(X_cat[['maker']]).toarray(), columns=ohe.get_feature_names())
X_combined=pd.concat([X_cat_transformed, X_numerical], axis=1)
X_combined
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X_combined, y)
display(X_train.shape)
display(X_test.shape)
display(y_train.shape)
display(y_test.shape)
```

```python
display(pd.concat([X_combined, y], axis=1)[['mileage', 'price_eur']].corr())
pd.concat([X_combined, y], axis=1)[['mileage', 'price_eur']].plot(kind='scatter', x='mileage', y='price_eur')
```

```python
rfr=RandomForestRegressor()
rfr.fit(X_train, y_train)
display(rfr.score(X_train, y_train))
display(rfr.score(X_test, y_test))
```
